## Histogram

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
f=plt.figure(figsize=(6,4))
gaussian_numbers = np.random.normal(size=10000)
histog = f.add_subplot(1,1,1)
histog.hist(gaussian_numbers, ec='black')
# plt.hist(gaussian_numbers, ec='black')
plt.title("Gaussian Histogram")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()
```

```python
n, bins, patches = plt.hist(gaussian_numbers, ec='black')
print("n: ", n, sum(n))
print("bins: ", bins)
for i in range(len(bins)-1):
    print(bins[i+1] -bins[i])
print("patches: ", patches)

plt.setp(patches[2], facecolor='g')
```

```python
plt.hist(gaussian_numbers, bins=100,ec='black')
plt.show()
```

```python
# If it is set to 'True', the first element of the return tuple will be the counts normalized
# to form a probability density, i.e., "n/(len(x)`dbin)", ie the integral of the histogram will
# sum to 1.    
plt.hist(gaussian_numbers, bins=100, normed=True,ec='black')
plt.show()
```

```python
# If both the parameters 'normed' and 'stacked' are set to 'True', the sum of the histograms is
# normalized to 1.
plt.hist(gaussian_numbers, 
         bins=100, 
         density=True, 
         stacked=True,
         histtype = 'step',
         edgecolor="#6A9662",
         color="#DDFFDD")
plt.show()
```

```python
# We can plot it as a cumulative distribution function as well by setting the parameter 
# 'cumulative'.
plt.hist(gaussian_numbers, 
         bins=100, 
         normed=True,
         stacked=True,
         cumulative=True,ec='black')
plt.show()
```

## Countour plot

```python
# A contour line or isoline of a function of two variables is a curve along which the function 
# has a constant value.
xlist = np.linspace(-3.0, 3.0, 3)
ylist = np.linspace(-3.0, 3.0, 4)
X, Y = np.meshgrid(xlist, ylist)
print(xlist)
print('\n')
print(ylist)
print('\n')
print(X)
print('\n')
print(Y)
plt.plot(X,Y,marker='.',linestyle='none')
```

```python
xlist = np.linspace(-3.0, 3.0, 3)
ylist = np.linspace(-3.0, 3.0, 4)
X, Y = np.meshgrid(xlist, ylist)
Z = np.sqrt(X**2 + Y**2)
print(Z)
```

```python
plt.figure()
cp = plt.contour(X, Y, Z)
plt.clabel(cp, inline=True,fontsize=10)
plt.title('Contour Plot')
plt.xlabel('x (cm)')
plt.ylabel('y (cm)')
plt.show()
```

```python
plt.figure()
cp = plt.contour(X, Y, Z, colors='black', linestyles='dashed')
plt.clabel(cp, inline=True, 
          fontsize=10)
plt.title('Contour Plot')
plt.xlabel('x (cm)')
plt.ylabel('y (cm)')
plt.show()
```

```python
#filled countour
xlist = np.linspace(-3.0, 3.0, 100)
ylist = np.linspace(-3.0, 3.0, 100)
X, Y = np.meshgrid(xlist, ylist)
Z = np.sqrt(X**2 + Y**2)
plt.figure()
cp = plt.contourf(X, Y, Z)
plt.colorbar(cp)
plt.title('Filled Contours Plot')
plt.xlabel('x (cm)')
plt.ylabel('y (cm)')
plt.show()
```

```python
xlist = np.linspace(-3.0, 3.0, 100)
ylist = np.linspace(-3.0, 3.0, 100)
X, Y = np.meshgrid(xlist, ylist)
Z = np.sqrt(X**2 + Y**2)
plt.figure()
contour = plt.contourf(X, Y, Z)
plt.clabel(contour, colors = 'k', fmt = '%2.1f', fontsize=12)
c = ('#ff0000', '#ffff00', '#0000FF', '0.6', 'c', 'm')
contour_filled = plt.contourf(X, Y, Z, colors=c)
plt.colorbar(contour_filled)
plt.title('Filled Contours Plot')
plt.xlabel('x (cm)')
plt.ylabel('y (cm)')
plt.show()
```

```python
#levels
X, Y = np.meshgrid(xlist, ylist)
Z = np.sqrt(X ** 2 + Y ** 2 )
plt.figure()
levels = [0.0, 0.2, 0.5, 0.9, 1.5, 2.5, 3.5]
contour = plt.contour(X, Y, Z, levels, colors='k')
plt.clabel(contour, colors = 'k', fmt = '%2.1f', fontsize=12)
contour_filled = plt.contourf(X, Y, Z, levels)
plt.colorbar(contour_filled)
plt.title('Plot from level list')
plt.xlabel('x (cm)')
plt.ylabel('y (cm)')
plt.show()
```
